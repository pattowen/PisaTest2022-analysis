```python
# Calculate the 25th and 75th percentiles based on all countries combined
global_percentiles = overall_score['overall_score'].quantile([0.25, 0.75])
bottom_25_threshold = global_percentiles[0.25]
top_25_threshold = global_percentiles[0.75]

# Display the global thresholds for performance groups
performance_stats = pd.DataFrame({
    "Statistic": ["25th Percentile (Bottom 25%)", "75th Percentile (Top 25%)"],
    "Value": [bottom_25_threshold, top_25_threshold]
})
```

```python
# Classify students globally without adding a new column
overall_score['Performance Group'] = overall_score['overall_score'].apply(
    lambda score: 'Bottom 25%' if score <= bottom_25_threshold else (
        'Top 25%' if score >= top_25_threshold else 'Middle 50%'
    )
)
overall_score.describe()
overall_score
```

```python


### Load QA data
questions = pd.read_csv(r"C:\1stApril\Work\4rd_Year\4rd2\CSS400_Project_Development\Raw_data\pisa2022.csv", encoding='utf-8')
questions.head()
```

```python
st034_df = questions[[ 
    "CNTSTUID", 
    "ST034Q01TA", 
    "ST034Q02TA", 
    "ST034Q03TA", 
    "ST034Q04TA", 
    "ST034Q05TA", 
    "ST034Q06TA"
]]

st034_df
column_renames = {
    "CNTSTUID" : 'Student ID',
    "ST034Q01TA": "I feel like an outsider at school",
    "ST034Q02TA": "I make friends easily at school",
    "ST034Q03TA": "I feel like I belong at school",
    "ST034Q04TA": "I feel awkward and out of place in my school",
    "ST034Q05TA": "Other students seem to like me",
    "ST034Q06TA": "I feel lonely at school"
}

st034_df= st034_df.rename(columns=column_renames)

value_mapping = {
    1: "Strongly agree",
    2: "Agree",
    3: "Disagree",
    4: "Strongly disagree"
}

# Apply the mapping to the renamed DataFrame
st034_df = st034_df.applymap(lambda x: value_mapping.get(x, x))
st034_df
# Rename the 'student_id' column in both DataFrames for consistency
overall_score.rename(columns={'student_id': 'Student ID', 'country': 'Country', 'overall_score': 'OVERALL SCORE'}, inplace=True)


# Assuming st034_df is already loaded and needs alignment (rename its column to 'Student ID')
st034_df.rename(columns={'student_id': 'Student ID'}, inplace=True)

# Merge the two DataFrames on 'Student ID'
merged_df = overall_score.merge(st034_df, on='Student ID', how='inner')
merged_df
merged_df.to_csv('s34.csv', index=False)
s34 = merged_df
# Define positive and negative questions
positive_questions = [
    "I make friends easily at school",
    "I feel like I belong at school",
    "Other students seem to like me"
]

negative_questions = [
    "I feel like an outsider at school",
    "I feel awkward and out of place in my school",
    "I feel lonely at school"
]

# Combine all response columns
response_columns = positive_questions + negative_questions

# Create a copy of the DataFrame to avoid modifying the original data
s34_numeric = s34.copy()

# Define a function to map responses to numeric values
def response_to_numeric(response):
    if response in ["Agree", "Strongly agree"]:
        return 1
    else:
        return 0

# Apply the mapping function to the response columns
s34_numeric[response_columns] = s34_numeric[response_columns].applymap(response_to_numeric)

# Extract positive and negative responses for visualization
positive_responses = s34_numeric[positive_questions].mean() * 100
negative_responses = s34_numeric[negative_questions].mean() * 100

# Define positions and bar width for visualization
positions = np.arange(len(positive_questions))
bar_width = 0.35

# Create the figure and subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 7), gridspec_kw={'width_ratios': [1, 1]})

# Plot positive responses on the left
axes[0].barh(positions, positive_responses.values, color='skyblue', alpha=0.8)
axes[0].set_yticks(positions)
axes[0].set_yticklabels(positive_questions, fontsize=12)
axes[0].invert_yaxis()
axes[0].set_xlim(0, 100)
axes[0].set_title("Positive Sense of Belonging", fontsize=14)
axes[0].set_xlabel("Agree or Strongly Agree (%)", fontsize=12)

# Add data labels
for i, v in enumerate(positive_responses.values):
    axes[0].text(v + 1, i, f"{v:.1f}%", va='center', fontsize=12)

# Plot negative responses on the right
axes[1].barh(positions, negative_responses.values, color='salmon', alpha=0.8)
axes[1].set_yticks(positions)
axes[1].set_yticklabels(negative_questions, fontsize=12)
axes[1].invert_yaxis()
axes[1].set_xlim(0, 100)
axes[1].set_title("Negative Sense of Belonging", fontsize=14)
axes[1].set_xlabel("Agree or Strongly Agree (%)", fontsize=12)

# Add data labels
for i, v in enumerate(negative_responses.values):
    axes[1].text(v + 1, i, f"{v:.1f}%", va='center', fontsize=12)

# Improve layout
plt.tight_layout(pad=3.0)
plt.show()


```

```python
# Filter data specifically for Thailand, Singapore, and OECD countries
categories = ["Thailand", "Singapore", "OECD"]
s34_numeric['Category'] = s34_numeric['Country']
s34_numeric.loc[s34_numeric['OECD Status'] == 'OECD', 'Category'] = 'OECD'

# Group data by the defined categories and calculate the means
grouped_data = s34_numeric[s34_numeric['Category'].isin(categories)].groupby('Category')[positive_questions + negative_questions].mean() * 100

# Define positions and bar width for visualization
positions = np.arange(len(positive_questions + negative_questions))
bar_width = 0.25
# Define separate positions for positive and negative questions
positive_positions = np.arange(len(positive_questions))
negative_positions = np.arange(len(negative_questions))

fig, axes = plt.subplots(1, 2, figsize=(16, 8), gridspec_kw={'width_ratios': [1, 1]})

# Plot positive responses
positive_grouped = grouped_data[positive_questions]

for i, category in enumerate(categories):
    positive_data = positive_grouped.loc[category]
    axes[0].barh(
        positive_positions + (i * bar_width),
        positive_data.values,
        height=bar_width,
        label=category
    )

axes[0].set_yticks(positive_positions + bar_width)
axes[0].set_yticklabels(positive_questions, fontsize=12)
axes[0].invert_yaxis()
axes[0].set_xlim(0, 100)
axes[0].set_title("Positive Sense of Belonging", fontsize=14)
axes[0].set_xlabel("Agree or Strongly Agree (%)", fontsize=12)

# Add data labels for positive questions
for i, category in enumerate(categories):
    positive_data = positive_grouped.loc[category]
    for j, value in enumerate(positive_data.values):
        axes[0].text(
            value + 1,
            positive_positions[j] + (i * bar_width),
            f"{value:.1f}%",
            va='center',
            fontsize=10
        )

# Plot negative responses
negative_grouped = grouped_data[negative_questions]

for i, category in enumerate(categories):
    negative_data = negative_grouped.loc[category]
    axes[1].barh(
        negative_positions + (i * bar_width),
        negative_data.values,
        height=bar_width,
        label=category
    )

axes[1].set_yticks(negative_positions + bar_width)
axes[1].set_yticklabels(negative_questions, fontsize=12)
axes[1].invert_yaxis()
axes[1].set_xlim(0, 100)
axes[1].set_title("Negative Sense of Belonging", fontsize=14)
axes[1].set_xlabel("Agree or Strongly Agree (%)", fontsize=12)

# Add data labels for negative questions
for i, category in enumerate(categories):
    negative_data = negative_grouped.loc[category]
    for j, value in enumerate(negative_data.values):
        axes[1].text(
            value + 1,
            negative_positions[j] + (i * bar_width),
            f"{value:.1f}%",
            va='center',
            fontsize=10
        )

# Add legend
axes[0].legend(title="Country", loc="lower right", fontsize=10)
axes[1].legend(title="Country", loc="lower right", fontsize=10)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
