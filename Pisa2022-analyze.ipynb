```python
code_to_label = {
    'CNTSTUID': 'Student ID',
    'CNTSCHID': 'School ID',
    'CNT': 'Country',
    'OECD': 'OECD Status',
    'ST004D01T': 'Gender',
    'AGE': 'Age',
    'ESCS': 'Economic, Social, and Cultural Status',
    'HISEI': 'Highest Parental Occupational Status (ISEI)',
    'FISCED': 'Father\'s level of education (ISCED)',
    'MISCED': 'Mother\'s level of education (ISCED)',
    'ST127Q01TA': 'Grade Repetition (ISCED 1)',
    'ST127Q02TA': 'Grade Repetition (ISCED 2)',
    'ST127Q03TA': 'Grade Repetition (ISCED 3)',
}

# Assuming pisa2022_df is the DataFrame with your data, we will rename the columns according to the mapping
student_info_df = pisa2022_df[[
    'CNTSTUID', 'CNTSCHID', 'CNT', 'OECD', 'ST004D01T', 'AGE', 'ESCS',
    'HISEI', 'FISCED','MISCED', 'ST127Q01TA', 'ST127Q02TA', 'ST127Q03TA'
]].rename(columns=code_to_label)
student_info_df
```

+ PV1MATH to PV10MATH are for mathematical score

+ PV1SCIE to PV10SCIE for scientific score

+ PV1READ to PV10READ for reading score

I will create 3 variables each of which will be created from average of 10 plausible variables found in each of mathematical, scientific, reading literacy, as shown below. The following 3 variables that will be created are called:

+ MATH SCORE

+ READING SCORE

+ SCIENCE SCORE

```python
# Calculate the Math Score
student_score_df['MATH SCORE'] = (
    student_score_df['PV1MATH'] + student_score_df['PV2MATH'] + student_score_df['PV3MATH'] +
    student_score_df['PV4MATH'] + student_score_df['PV5MATH'] + student_score_df['PV6MATH'] +
    student_score_df['PV7MATH'] + student_score_df['PV8MATH'] + student_score_df['PV9MATH'] +
    student_score_df['PV10MATH']
) / 10

# Calculate the Reading Score
student_score_df['READING SCORE'] = (
    student_score_df['PV1READ'] + student_score_df['PV2READ'] + student_score_df['PV3READ'] +
    student_score_df['PV4READ'] + student_score_df['PV5READ'] + student_score_df['PV6READ'] +
    student_score_df['PV7READ'] + student_score_df['PV8READ'] + student_score_df['PV9READ'] +
    student_score_df['PV10READ']
) / 10

# Calculate the Science Score
student_score_df['SCIENCE SCORE'] = (
    student_score_df['PV1SCIE'] + student_score_df['PV2SCIE'] + student_score_df['PV3SCIE'] +
    student_score_df['PV4SCIE'] + student_score_df['PV5SCIE'] + student_score_df['PV6SCIE'] +
    student_score_df['PV7SCIE'] + student_score_df['PV8SCIE'] + student_score_df['PV9SCIE'] +
    student_score_df['PV10SCIE']
) / 10

# List of columns to drop
columns_to_drop = [
    'PV1MATH', 'PV2MATH', 'PV3MATH', 'PV4MATH', 'PV5MATH', 'PV6MATH', 'PV7MATH', 'PV8MATH', 'PV9MATH', 'PV10MATH',
    'PV1READ', 'PV2READ', 'PV3READ', 'PV4READ', 'PV5READ', 'PV6READ', 'PV7READ', 'PV8READ', 'PV9READ', 'PV10READ',
    'PV1SCIE', 'PV2SCIE', 'PV3SCIE', 'PV4SCIE', 'PV5SCIE', 'PV6SCIE', 'PV7SCIE', 'PV8SCIE', 'PV9SCIE', 'PV10SCIE'
]

# Drop the used columns
student_score_df.drop(columns=columns_to_drop, inplace=True)

```

```python
student_score_df['OVERALL SCORE'] = student_score_df[['MATH SCORE','SCIENCE SCORE','READING SCORE']].mean(axis=1)
student_score_df
```

```python
# Group data by parental education levels and calculate mean scores
parental_education_impact = thai_students.groupby(
    ["Father's level of education (ISCED)", "Mother's level of education (ISCED)"]
)[['MATH SCORE', 'READING SCORE', 'SCIENCE SCORE', 'OVERALL SCORE']].mean().reset_index()


fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for ax, subject, title in zip(axes.flatten(), subjects, titles):
    pivot_data = parental_education_impact.pivot(
        index="Father's level of education (ISCED)", 
        columns="Mother's level of education (ISCED)", 
        values=subject
    )
    heatmap = ax.imshow(pivot_data, cmap='viridis', aspect='auto')
    ax.set_title(title, fontsize=12)
    ax.set_xlabel("Mother's Level of Education (ISCED)", fontsize=10)
    ax.set_ylabel("Father's Level of Education (ISCED)", fontsize=10)
    ax.set_xticks(range(len(pivot_data.columns)))
    ax.set_xticklabels(pivot_data.columns, rotation=45)
    ax.set_yticks(range(len(pivot_data.index)))
    ax.set_yticklabels(pivot_data.index)
    fig.colorbar(heatmap, ax=ax, orientation='vertical', label='Average Score')

plt.tight_layout()
plt.show()

```

From above paragraph: \
Higher levels of parental education are strongly linked to better student performance across all subjects, with ISCED levels 5 or 6 (higher education) yielding the highest scores. The combined education level of both parents enhances outcomes, though the mother's education shows a slightly stronger correlation with Reading scores, likely due to early literacy support. Students with parents at ISCED levels 0 or 1 (minimal education) show significantly lower scores, highlighting the need for targeted interventions, such as parent education programs and literacy initiatives, to bridge these disparities.

```python
occupational_and_economic_impact = thai_students.groupby(
    ['Highest Parental Occupational Status (ISEI)', 'Economic, Social, and Cultural Status']
)[['MATH SCORE', 'READING SCORE', 'SCIENCE SCORE', 'OVERALL SCORE']].mean().reset_index()

# Filter out rows where 'Highest Parental Occupational Status (ISEI)' is 0
filtered_data = occupational_and_economic_impact[
    occupational_and_economic_impact['Highest Parental Occupational Status (ISEI)'] != 0
]

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for ax, subject, title in zip(axes.flatten(), subjects, titles):
    scatter = ax.scatter(
        filtered_data['Highest Parental Occupational Status (ISEI)'],
        filtered_data[subject],
        c=filtered_data['Economic, Social, and Cultural Status'],
        cmap='coolwarm',
        edgecolor='k',
        alpha=0.7
    )
    ax.set_title(title, fontsize=12)
    ax.set_xlabel('Parental Occupational Status (ISEI)', fontsize=10)
    ax.set_ylabel('Average Score', fontsize=10)
    fig.colorbar(scatter, ax=ax, label='ESCS')

plt.tight_layout()
plt.show()
```

From these graph: \
The analysis reveals that higher parental occupational status (ISEI) is strongly associated with better student performance across all subjects, with the trend being more pronounced in Mathematics and Science. Economic, Social, and Cultural Status (ESCS) further amplifies this impact, as students from families with higher ESCS consistently achieve higher scores. The scatterplots indicate a positive gradient, where better occupational and economic conditions align with superior academic outcomes, underscoring the importance of socioeconomic resources in fostering educational success.
